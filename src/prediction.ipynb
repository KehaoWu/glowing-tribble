```python
import tensorflow as tf
```

```python
print("Current tensorflow version is {}".format(tf.__version__))
```

```python
len_sentence = 100
```

```python
len_word2vec = 200
```

```python
size_rnn = 4
```

```python
num_layers = 1
```

```python
input_x = tf.placeholder(tf.float32, [None, len_sentence, len_word2vec])
```

```python
fw_cell = tf.nn.rnn_cell.LSTMCell(size_rnn, state_is_tuple=True)
```

```python
fw_cell = tf.nn.rnn_cell.DropoutWrapper(fw_cell, output_keep_prob=0.5)
```

```python
bw_cell = tf.nn.rnn_cell.LSTMCell(size_rnn, state_is_tuple=True)
```

```python
bw_cell = tf.nn.rnn_cell.DropoutWrapper(bw_cell, output_keep_prob=0.5)
```

```python
fw_cell = tf.nn.rnn_cell.MultiRNNCell([fw_cell] * num_layers, state_is_tuple=True)
```

```python
bw_cell = tf.nn.rnn_cell.MultiRNNCell([bw_cell] * num_layers, state_is_tuple=True)
```

```python
words_used_in_sent = tf.sign(tf.reduce_max(tf.abs(input_x), reduction_indices=2))
```

```python
length = tf.cast(tf.reduce_sum(words_used_in_sent, reduction_indices=1), tf.int32)
```

```python
def weight_and_bias(in_size, out_size):
    weight = tf.truncated_normal([in_size, out_size], stddev=0.01)
    bias = tf.constant(0.1, shape=[out_size])
    return tf.Variable(weight), tf.Variable(bias)
```

```python
output, _, _ = tf.nn.static_bidirectional_rnn(fw_cell, bw_cell,
                                             tf.unstack(tf.transpose(input_x, perm=[1, 0, 2])),
                                             dtype=tf.float32, sequence_length=length)
```

```python
size_rnn = 4
```

```python
size_output = 4
```

```python
weight, bias = weight_and_bias(2 * size_rnn, size_output)
```

```python
output = tf.reshape(tf.transpose(tf.stack(output), perm=[1, 0, 2]), [-1, 2 * size_rnn])
```

```python
prediction = tf.nn.softmax(tf.matmul(output, weight) + bias)
```

```python
prediction = tf.reshape(prediction, [-1, len_sentence, size_output])
```
